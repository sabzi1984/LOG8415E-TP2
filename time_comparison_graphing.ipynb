```python
import pandas as pd
df_spark_time=pd.read_csv('/content/drive/MyDrive/Automn 2021/Advanced Cloud Computin/user time/spark_time.csv')
df_hadoop_time=pd.read_csv('/content/drive/MyDrive/Automn 2021/Advanced Cloud Computin/user time/hadoop_time_new.csv')
df_linux_time=pd.read_csv('/content/drive/MyDrive/Automn 2021/Advanced Cloud Computin/user time/linux_time.csv')
```

```python
import re
import pandas as pd
df_spark_time['real_time']=df_spark_time['real'].str.findall(r'\d\.\d+').apply(lambda x: " ".join(x))
df_spark_time['user_time']=df_spark_time['user'].str.findall(r'\d\.\d+').apply(lambda x: " ".join(x))
df_spark_time['sys_time']=df_spark_time['sys'].str.findall(r'\d\.\d+').apply(lambda x: " ".join(x))
df_spark_time['real_time'] = pd.to_numeric(df_spark_time['real_time'])
df_spark_time['user_time'] = pd.to_numeric(df_spark_time['user_time'])
df_spark_time['sys_time'] = pd.to_numeric(df_spark_time['sys_time'])
# df_spark_time['real_1']=float(df_spark_time['real_1'])
# df_spark_time['user_1']=float(df_spark_time['user_1'])
# df_spark_time['sys_1']=float(df_spark_time['sys_1'])
df_spark_time
```

```python
import re
  
df_hadoop_time['real_time']=df_hadoop_time['real'].str.findall(r'\d\.\d+').apply(lambda x: " ".join(x))
df_hadoop_time['user_time']=df_hadoop_time['user'].str.findall(r'\d\.\d+').apply(lambda x: " ".join(x))
df_hadoop_time['sys_time']=df_hadoop_time['sys'].str.findall(r'\d\.\d+').apply(lambda x: " ".join(x))
df_hadoop_time['real_time'] = pd.to_numeric(df_hadoop_time['real_time'])
df_hadoop_time['user_time'] = pd.to_numeric(df_hadoop_time['user_time'])
df_hadoop_time['sys_time'] = pd.to_numeric(df_hadoop_time['sys_time'])

df_hadoop_time
```

```python
import re
  
df_linux_time['real_time']=df_linux_time['real'].str.findall(r'\d\.\d+').apply(lambda x: " ".join(x))
df_linux_time['user_time']=df_linux_time['user'].str.findall(r'\d\.\d+').apply(lambda x: " ".join(x))
df_linux_time['sys_time']=df_linux_time['sys'].str.findall(r'\d\.\d+').apply(lambda x: " ".join(x))
df_linux_time['real_time'] = pd.to_numeric(df_linux_time['real_time'])
df_linux_time['user_time'] = pd.to_numeric(df_linux_time['user_time'])
df_linux_time['sys_time'] = pd.to_numeric(df_linux_time['sys_time'])

df_linux_time
```

```python
df_spark_time_process=df_spark_time.groupby(df_spark_time['Book']).agg({'real_time':['mean'],'user_time':['mean'], 'sys_time':['mean']})
df_spark_time_process.columns=['mean real time', 'mean user time', 'mean sys time']
df_spark_time_process.reset_index(inplace=True)
df_spark_time_process
```

```python
df_hadoop_time_process=df_hadoop_time.groupby(df_hadoop_time['Book']).agg({'real_time':['mean'],'user_time':['mean'], 'sys_time':['mean']})
df_hadoop_time_process.columns=['mean real time', 'mean user time', 'mean sys time']
df_hadoop_time_process.reset_index(inplace=True)
df_hadoop_time_process
```

```python
df_linux_time_process=df_linux_time.groupby(df_linux_time['Book']).agg({'real_time':['mean'],'user_time':['mean'], 'sys_time':['mean']})
df_linux_time_process.columns=['mean real time', 'mean user time', 'mean sys time']
df_linux_time_process.reset_index(inplace=True)
df_linux_time_process
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Bar(
    x=df_spark_time_process['Book'],
    y=df_spark_time_process['mean user time'],
    name='spark user time',
    marker_color='#bc5090'
))
fig.add_trace(go.Bar(
     x=df_hadoop_time_process['Book'],
    y=df_hadoop_time_process['mean user time'],
    name='hadoop user time',
    marker_color='#58508d'
))



fig.show()
```
